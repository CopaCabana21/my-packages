```python
from concurrent.futures import ThreadPoolExecutor, TimeoutError

def my_function(x):
    import time
    time.sleep(x)  # Simulate varying execution time
    return f"Processed {x}"

def tryFunction(function, x, timeout=3):
    with ThreadPoolExecutor(max_workers=1) as executor:
        future = executor.submit(function, x)
        try:
            print(f"processing: {x} ...")
            result = future.result(timeout=timeout)
            print("finished: success")
            return result # Set a timeout for each call
        except TimeoutError:
            print(f"finished: timeout")
            return 0

# List of inputs to process
inputs = [1, 2, 5, 3]

# Apply the function with timeouts
results = [tryFunction(my_function,x) for x in inputs]
print(results)
```
